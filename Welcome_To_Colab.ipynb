```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def solve_laplace_mixedBC_vec(Nx, Ny, Lx=1.0, Ly=1.0, tol=1e-5, max_iter=10000):
    dx, dy = Lx/(Nx-1), Ly/(Ny-1)
    u = np.full((Nx, Ny), 50.0)

    # Dirichlet BC
    u[0, :]  = 100.0  # left
    u[-1, :] = 0.0    # right

    for it in range(max_iter):
        u_old = u.copy()

        # Neumann BC (top/bottom insulated)
        u[:, 0]  = u[:, 1]
        u[:, -1] = u[:, -2]

        # Vectorized update for interior points
        u[1:-1, 1:-1] = 0.25 * (u[2:, 1:-1] + u[:-2, 1:-1] + u[1:-1, 2:] + u[1:-1, :-2])

        # Reapply Dirichlet BC
        u[0, :]  = 100.0
        u[-1, :] = 0.0

        # Convergence check
        if np.max(np.abs(u - u_old)) < tol:
            print(f"Converged after {it} iterations for {Nx}x{Ny}")
            break

    return u

# Solve for 50, 100, 500 grids
grids = [50, 100, 500]
solutions = []

for N in grids:
    u = solve_laplace_mixedBC_vec(N, N)
    solutions.append(u)

# Plot contour and surface plots
for N, u in zip(grids, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))

    # Contour plot
    plt.figure(figsize=(6,5))
    plt.contourf(X, Y, u.T, 50, cmap='inferno')
    plt.title(f'Contour Plot {N}x{N}')
    plt.xlabel('x'); plt.ylabel('y')
    plt.colorbar(label='u(x,y)')
    plt.show()

    # Surface plot
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, u.T, cmap='inferno')
    ax.set_title(f'Surface Plot {N}x{N}')
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u(x,y)')
    plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def jacobi_laplace(Nx, Ny, Lx=1.0, Ly=1.0, tol=1e-5, max_iter=10000):
    """
    Solve 2D Laplace equation with mixed BCs using Jacobi iteration
    Dirichlet on left/right, Neumann (insulated) on top/bottom.
    """
    u = np.full((Nx, Ny), 50.0)

    # Dirichlet BC
    u[0, :] = 100.0
    u[-1, :] = 0.0

    for it in range(max_iter):
        u_old = u.copy()

        # Neumann BC (top/bottom)
        u[:, 0] = u[:, 1]
        u[:, -1] = u[:, -2]

        # Jacobi update for interior
        u[1:-1, 1:-1] = 0.25 * (u_old[2:, 1:-1] + u_old[:-2, 1:-1] +
                                  u_old[1:-1, 2:] + u_old[1:-1, :-2])

        # Reapply Dirichlet BC
        u[0, :] = 100.0
        u[-1, :] = 0.0

        # Check convergence
        if np.max(np.abs(u - u_old)) < tol:
            print(f"Jacobi converged after {it} iterations for {Nx}x{Ny}")
            break

    return u

# Grid sizes
grid_sizes = [50, 100, 500]

# Solve
solutions = []
for N in grid_sizes:
    u = jacobi_laplace(N, N)
    solutions.append(u)

# -------------------------
# Contour Plots
# -------------------------
for N, u in zip(grid_sizes, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    plt.figure(figsize=(6,5))
    contour = plt.contourf(X, Y, u.T, 50, cmap='viridis')
    plt.title(f'Jacobi Contour Plot {N}x{N}')
    plt.xlabel('x'); plt.ylabel('y')
    plt.colorbar(contour, label='u(x,y)')
    plt.show()

# -------------------------
# Surface Plots
# -------------------------
for N, u in zip(grid_sizes, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, u.T, cmap='viridis')
    ax.set_title(f'Jacobi Surface Plot {N}x{N}')
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u(x,y)')
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
    plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time

def sor_laplace(Nx, Ny, Lx=1.0, Ly=1.0, tol=1e-5, max_iter=10000, omega=1.7):
    """
    Solve 2D Laplace equation with mixed BCs using SOR iteration
    Dirichlet on left/right, Neumann (insulated) on top/bottom.
    Returns solution, number of iterations, elapsed time
    """
    u = np.full((Nx, Ny), 50.0)

    # Dirichlet BC
    u[0, :] = 100.0
    u[-1, :] = 0.0

    start_time = time.time()
    for it in range(max_iter):
        u_old = u.copy()

        # Neumann BC (top/bottom)
        u[:, 0] = u[:, 1]
        u[:, -1] = u[:, -2]

        # SOR update for interior points
        for i in range(1, Nx-1):
            for j in range(1, Ny-1):
                u[i, j] = (1-omega)*u[i, j] + omega*0.25*(u[i+1,j] + u[i-1,j] + u[i,j+1] + u[i,j-1])

        # Reapply Dirichlet BC
        u[0, :] = 100.0
        u[-1, :] = 0.0

        # Check convergence
        if np.max(np.abs(u - u_old)) < tol:
            elapsed = time.time() - start_time
            print(f"SOR converged after {it+1} iterations for {Nx}x{Ny}, Time: {elapsed:.2f} sec")
            return u, it+1, elapsed

    elapsed = time.time() - start_time
    print(f"SOR reached max iterations ({max_iter}) for {Nx}x{Ny}, Time: {elapsed:.2f} sec")
    return u, max_iter, elapsed

# Grid sizes
grid_sizes = [50, 100, 500]

# Solve and store results
solutions = []
for N in grid_sizes:
    u, iterations, elapsed = sor_laplace(N, N, omega=1.7)
    solutions.append((u, iterations, elapsed))

# -------------------------
# Contour Plots
# -------------------------
for (N, (u, iterations, elapsed)) in zip(grid_sizes, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    plt.figure(figsize=(6,5))
    contour = plt.contourf(X, Y, u.T, 50, cmap='plasma')
    plt.title(f'SOR Contour Plot {N}x{N} (Iterations: {iterations}, Time: {elapsed:.2f}s)')
    plt.xlabel('x'); plt.ylabel('y')
    plt.colorbar(contour, label='u(x,y)')
    plt.show()

# -------------------------
# Surface Plots
# -------------------------
for (N, (u, iterations, elapsed)) in zip(grid_sizes, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, u.T, cmap='plasma')
    ax.set_title(f'SOR Surface Plot {N}x{N} (Iterations: {iterations}, Time: {elapsed:.2f}s)')
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u(x,y)')
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
    plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def gauss_seidel_laplace(Nx, Ny, Lx=1.0, Ly=1.0, tol=1e-5, max_iter=10000):
    """
    Solve 2D Laplace equation using GaussSeidel iteration
    Dirichlet on left/right, Neumann (insulated) on top/bottom.
    """
    u = np.full((Nx, Ny), 50.0)

    # Dirichlet BC
    u[0, :] = 100.0
    u[-1, :] = 0.0

    for it in range(max_iter):
        u_old = u.copy()

        # Apply Neumann BCs (top/bottom)
        u[:, 0] = u[:, 1]
        u[:, -1] = u[:, -2]

        # Gauss-Seidel update (use in-place updates)
        for i in range(1, Nx-1):
            for j in range(1, Ny-1):
                u[i, j] = 0.25 * (u[i+1, j] + u[i-1, j] +
                                  u[i, j+1] + u[i, j-1])

        # Reapply Dirichlet BCs
        u[0, :] = 100.0
        u[-1, :] = 0.0

        # Check convergence
        if np.max(np.abs(u - u_old)) < tol:
            print(f"GaussSeidel converged after {it} iterations for {Nx}x{Ny}")
            break

    return u


# Test on multiple grid sizes
grid_sizes = [50, 100, 500]
solutions = []

for N in grid_sizes:
    print(f"\nSolving {N}x{N} using GaussSeidel...")
    u = gauss_seidel_laplace(N, N)
    solutions.append(u)

# -------------------------
# Contour Plots
# -------------------------
for N, u in zip(grid_sizes, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    plt.figure(figsize=(6,5))
    contour = plt.contourf(X, Y, u.T, 50, cmap='plasma')
    plt.title(f'GaussSeidel Contour Plot {N}x{N}')
    plt.xlabel('x'); plt.ylabel('y')
    plt.colorbar(contour, label='u(x,y)')
    plt.show()

# -------------------------
# Surface Plots
# -------------------------
for N, u in zip(grid_sizes, solutions):
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, u.T, cmap='plasma')
    ax.set_title(f'GaussSeidel Surface Plot {N}x{N}')
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u(x,y)')
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
    plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import time
from mpl_toolkits.mplot3d import Axes3D

def gauss_seidel_laplace(Nx, Ny, Lx=1.0, Ly=1.0, tol=1e-5, max_iter=10000):
    """
    Solve 2D Laplace equation using GaussSeidel iteration
    Dirichlet BC: left=100, right=0
    Neumann BC: top & bottom insulated (du/dy=0)
    """
    dx, dy = Lx/(Nx-1), Ly/(Ny-1)
    u = np.full((Nx, Ny), 50.0)

    # Dirichlet BC
    u[0, :] = 100.0
    u[-1, :] = 0.0

    start_time = time.time()
    for it in range(max_iter):
        u_old = u.copy()

        # Apply Neumann BC (top and bottom)
        u[:, 0]  = u[:, 1]
        u[:, -1] = u[:, -2]

        # GaussSeidel update
        for i in range(1, Nx-1):
            for j in range(1, Ny-1):
                u[i, j] = 0.25 * (u[i+1, j] + u[i-1, j] +
                                  u[i, j+1] + u[i, j-1])

        # Reapply Dirichlet BC
        u[0, :] = 100.0
        u[-1, :] = 0.0

        # Check convergence
        if np.max(np.abs(u - u_old)) < tol:
            elapsed = time.time() - start_time
            print(f"GaussSeidel converged in {it} iterations ({elapsed:.2f} s) for {Nx}x{Ny}")
            break
    else:
        elapsed = time.time() - start_time
        print(f"Did not fully converge after {max_iter} iterations ({elapsed:.2f} s) for {Nx}x{Ny}")

    return u, it, elapsed


# ------------------------------
# Run for different grid sizes
# ------------------------------
grid_sizes = [50, 100, 500]
results = {}

for N in grid_sizes:
    print(f"\nSolving for {N}x{N} grid...")
    u, iterations, runtime = gauss_seidel_laplace(N, N)
    results[N] = {"u": u, "iterations": iterations, "runtime": runtime}


# ------------------------------
# Plot contour and surface for each grid
# ------------------------------
for N in grid_sizes:
    u = results[N]["u"]
    X, Y = np.meshgrid(np.linspace(0, 1, N), np.linspace(0, 1, N))

    # --- Contour Plot ---
    plt.figure(figsize=(6,5))
    contour = plt.contourf(X, Y, u.T, 50, cmap='plasma')
    plt.colorbar(contour, label='u(x,y)')
    plt.title(f"GaussSeidel Contour Plot ({N}{N})")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.tight_layout()
    plt.show()

    # --- Surface Plot ---
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, u.T, cmap='plasma')
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
    ax.set_title(f"GaussSeidel Surface Plot ({N}{N})")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("u(x,y)")
    plt.tight_layout()
    plt.show()


# ------------------------------
# Print summary table
# ------------------------------
print("\nSummary of GaussSeidel performance:\n")
print(f"{'Grid Size':>10} | {'Iterations':>10} | {'Time (s)':>10}")
print("-"*35)
for N in grid_sizes:
    print(f"{N:>10} | {results[N]['iterations']:>10} | {results[N]['runtime']:>10.3f}")
```

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time

def solve_laplace_sor(Nx, Ny, Lx=1.0, Ly=1.0, tol=1e-5, max_iter=30000, omega=1.5):
    dx, dy = Lx / (Nx - 1), Ly / (Ny - 1)
    u = np.zeros((Nx, Ny))

    # Dirichlet BC (left/right)
    u[0, :]  = 100.0
    u[-1, :] = 0.0

    for it in range(max_iter):
        u_old = u.copy()

        # Neumann BC (top/bottom)
        u[:, 0]  = u[:, 1]  # bottom insulated
        u[:, -1] = u[:, -2] # top insulated

        # SOR update for interior points
        for i in range(1, Nx-1):
            for j in range(1, Ny-1):
                u[i, j] = u[i, j] + omega * 0.25 * (
                    u[i+1, j] + u[i-1, j] + u[i, j+1] + u[i, j-1] - 4*u[i, j]
                )

        # Reapply Dirichlet BC
        u[0, :] = 100.0
        u[-1, :] = 0.0

        # Check convergence
        err = np.max(np.abs(u - u_old))
        if err < tol:
            return u, it

    return u, max_iter

# Run for multiple grid sizes
grid_sizes = [50, 100, 500]
results = {}

for N in grid_sizes:
    print(f"\nSolving {N}x{N} grid (SOR with omega={1.98})...")
    start = time.time()
    u, iters = solve_laplace_sor(N, N, omega=1.988)
    end = time.time()
    print(f"  Converged in {iters} iterations, time = {end - start:.2f}s")
    results[N] = (u, iters, end - start)

# Plot contours (50x50, 100x100, 500x500)
fig, axs = plt.subplots(1, 3, figsize=(15,5))
for ax, N in zip(axs, [50, 100, 500]):
    u = results[N][0]
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    c = ax.contourf(X, Y, u.T, 50, cmap='viridis')
    ax.set_title(f"SOR - {N}x{N}")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    fig.colorbar(c, ax=ax, label='Temperature')

plt.tight_layout()
plt.show()

# Plot surface plots (50x50, 100x100, 500x500)
fig = plt.figure(figsize=(18, 5))
for i, N in enumerate([50, 100, 500]):
    u = results[N][0]
    X, Y = np.meshgrid(np.linspace(0,1,N), np.linspace(0,1,N))
    ax = fig.add_subplot(1, 3, i+1, projection='3d')
    surf = ax.plot_surface(X, Y, u.T, cmap='viridis')
    ax.set_title(f"SOR - Surface {N}x{N}")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('Temperature')
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.tight_layout()
plt.show()

# Print summary
print("\nSummary:")
for N in grid_sizes:
    _, iters, t = results[N]
    print(f"{N:4d} x {N:4d} : {iters:6d} iterations, time = {t:6.2f} s")
```
